```python
from sqlalchemy import create_engine
import pandas as pd

# Define connection parameters
dbname = DATABASE_CONFIG['dbname']
user = DATABASE_CONFIG['user']
password = DATABASE_CONFIG['password']
host = DATABASE_CONFIG['host']
port = DATABASE_CONFIG['port']

# Create an engine
engine = create_engine(f"postgresql://{user}:{password}@{host}:{port}/{dbname}")
```

```python
pd.read_sql_query('''select * from survey.dim_sample''', con=engine)
```

```python
import json
```

```python
def point_to_json(point):
    return json.dumps({
        'type': 'Point',
        'coordinates': [point['x'], point['y']],
        'spatialReference': {'wkid': point['spatialReference']['wkid']}
    })
```

```python
list_df[1]['shape'] = list_df[1]['shape'].apply(point_to_json)
```

```python
list_df[0] = list_df[0].drop('shape', axis=1)
```

```python
list_df[1].to_sql(name = "dim_sample",con = engine, schema = "survey", if_exists = 'replace', index=False)
```
